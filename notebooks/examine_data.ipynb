```python
def print_datapoint(dataset, idx):
    print(f"EXAMPLE {idx}")
    print("QUERY: ", dataset['train'][idx]['query'])
    print("POSITIVE DOCUMENT: \n", dataset['train'][idx]['positive'].replace("\n", " "))
    print("NEGATIVE DOCUMENT: \n", dataset['train'][idx]['negative'].replace("\n", " "))
    print("--------------------------------")
```

```python
for i in range(0, 50, 5):
    print_datapoint(absp1train, i)
```
